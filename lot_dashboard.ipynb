### In order to assign IMDRF codes to the events table we need to transform the string type into a list with individual values.

```python
def make_list(row_val):
    
    row_val = row_val.replace('[', '')
    row_val = row_val.replace(']', '')
    if ',' in row_val:
        return row_val.split(',')
    else:
        return [row_val]
```

```python
merged['product_problems_clean'] = merged.product_problems.apply(make_list)
```

```python
# Iterate over each product_problems list in the merged DataFrame
imdrf_codes = []
for product_problems in merged['product_problems_clean']:
    matching_codes = []

    # Check each value in the product_problems list
    for problem in product_problems:
        # Check if the problem value is in the event_problem column of codes DataFrame
        matching_rows = codes[codes['event_problem'] == problem]

        # Get the corresponding imdrf_code values
        matching_codes.extend(matching_rows['imdrf_code'].tolist())

    imdrf_codes.append(matching_codes)

# Add the imdrf_codes list as a new column in the merged DataFrame
merged['imdrf_codes'] = imdrf_codes
```

### Lot Scope Dashboard has `is_scope` column as requirement. In order to evaluate if the lot under review is within the scope of recall we need to check if the list of `model_numbers` and `lot_numbers` from the `recall` table matches the `event` table, the difference will be `out_of_scope`

```python
is_scope_model = []
is_scope_lot = []
for _, row in merged.iterrows():
    if row['model_number'] in row['unpacked_model_lot']:
        is_scope_model.append('Yes')
    else:
        is_scope_model.append('No')
    
    if row['lot_number'] in row['unpacked_model_lot']:
        is_scope_lot.append('Yes')
    else:
        is_scope_lot.append('No')
```

```python
merged['is_scope_model'] = is_scope_model
merged['is_scope_lot'] = is_scope_lot
```
